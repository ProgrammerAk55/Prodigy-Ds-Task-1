#Extraction of top 20 countries with respect to total male population

```python
#Filter data for total male population
total_population_data=df[df['Indicator Code']=='SP.POP.TOTL.MA.IN']

#Sort data based on the total male population for 2024
total_population_sorted = total_population_data.sort_values(by='2024',ascending=False)

#Get the top twenty countries with the highest total male population for 2024
total_top_twenty_countries=total_population_sorted.head(20)

print(" Top 20 countries of total male population \n")
print(total_top_twenty_countries[['Country Name', 'Country Code']])
```

#Bar Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
# create a bar plot for 2024
plt.figure(figsize=(15,6))
sns.barplot(x="Country Code", y="2024", data=total_top_twenty_countries, hue="Country Code", palette="viridis", legend=False)
plt.title("Top 20 countries of total male population in year 2024",fontsize=10)
plt.xlabel("Country",fontsize=10)
plt.ylabel("Population",fontsize=10)
plt.show()
```

#A Decadal Overview of Male Population Distribution in India

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is already loaded
# df = pd.read_excel("your_file.xls", sheet_name="Data", engine="xlrd")

# ---------- 1. Bar Chart: Population of a Specific Country Over Time ----------
country_name = "India"
all_years = [str(year) for year in range(1960, 2024)]
decade_years = [str(year) for year in range(1960, 2024, 10)]  # every 10 years

country_data = df[df['Country Name'] == country_name]
if not country_data.empty:
    population = country_data[all_years].values.flatten()

    plt.figure(figsize=(12, 6))
    plt.bar(all_years, population, color='blue')
    plt.xticks(ticks=[i for i, y in enumerate(all_years) if y in decade_years],
               labels=[y for y in all_years if y in decade_years],
               rotation=45)
    plt.title(f'Population of {country_name} (1960–2024)')
    plt.xlabel('Year')
    plt.ylabel('Population')
    plt.tight_layout()
    plt.show()
else:
    print(f"Country '{country_name}' not found in the dataset.")
plt.show()
```

#Population Growth by Decade for All Countries

#Graph plot

```python
import pandas as pd
import plotly.graph_objs as go
import os # Import the os module
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('/content/WorldPopulationData.csv')

# Define decade columns from 1960 to 2020
decade_years = [str(year) for year in range(1960, 2025, 10)]

# Create Plotly figure
fig = go.Figure()

# Add a line for each country
for _, row in data.iterrows():
    country = row['Country Name']
    population = row[decade_years].values
    fig.add_trace(go.Scatter(
        x=decade_years,
        y=population,
        mode='lines+markers',
        name=country,
        hovertemplate=f"{country}Year: %{x}Population: %{y}",
        line=dict(width=1)
    ))

fig.update_layout(
    title="Male Population Growth by Decade (1960–2020) for All Countries",
    xaxis_title="Year",
    yaxis_title="Male Population",
    width=900,
    height=800,
    showlegend=False  # Set to True if you want to display all countries in legend
)
# Create the 'images' directory if it doesn't exist
if not os.path.exists('images'):
    os.makedirs('images')
fig.show()
plt.show()
```
